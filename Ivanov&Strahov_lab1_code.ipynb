```python
def radix_argsort(A):
    length = len(str(max(A)))
    deg = 10
    for i in range(length):
        B = [[] for k in range(deg)] 
        for x in A:
            B[(x // 10**i) % 10].append(x)
        A = []
        for k in range(deg):
            for x in B[k]:
                A.append(x)
    return A
```

```python
def insertionSort(A):
    for i in range (1, len(A)):
        t = A[i]
        j = i - 1
        while (j >= 0 and t < A[j]):
            A[j + 1] = A[j]
            j = j - 1
        A[j + 1] = t
    return A   

def partition(A, st, fn,pivot):
    A[pivot], A[fn] = A[fn], A[pivot]
    pivotValue = A[fn]
    i = st - 1
    for j in range(st, fn):
        if A[j] <= pivotValue:
            i += 1
            A[i], A[j] = A[j], A[i]  
    newPivot = i + 1
    A[newPivot], A[fn] = A[fn], A[newPivot]
    return newPivot
from random import randint
def quicksort(A, st, fn):
    while (fn-st>=1):
        pivot = randint(st, fn)
        pivot = partition(A, st, fn,pivot)
        if(pivot-st<=fn-pivot-1):
            quicksort(A, st, pivot)
            st=pivot+1
        else:
            quicksort(A, pivot + 1, fn)
            fn=pivot
    A[st:fn] = insertionSort(A[st:fn])

def quick_argsort(A):
    quicksort(A, 0, len(A)-1)
    return A
```

```python
visited = set()
ent={}
out={}
result=list()

def DFS(graph, v, parent, time):
    if v not in visited:
        visited.add(v)
    time = time + 1
    ent[v] = time
    out[v] = time
    for u in graph[v]:
        if (u==parent):
            continue
        if (u in visited):
            out[v] = min(out[v], ent[u])
        else:
            DFS(graph, u, v, time)
            out[v] = min(out[v], ent[u]) 
            if (out[u] > ent[v]):
                bridge = (v, u)
                result.append(tuple(sorted(bridge)))


def compute_bridges_determ(graph):
    result.clear()
    for i in graph:
        if (i not in visited):
            DFS(graph, i, -1, 0)
    visited.clear()
    ent.clear()
    out.clear()
    return result
```

```python
list_marks = defaultdict(list)

def dfs_rand(graph, v, parent, time):
    if v not in visited:
        visited.add(v)
    time = time + 1
    ent[v] = time
    out[v] = time
    for u in graph[v]:
        if (u==parent):
            continue
        if (u in visited):
            out[v] = min(out[v], ent[u])
        else:
            dfs_rand(graph, u, v, time)
            
            if len(graph[u]) == 1:
                list_marks[(u, v)] = 0
                list_marks[(v, u)] = 0
            else:
                a = 0
                for j in graph[u]:
                    if (u, j) in list_marks:
                        a = a ^ list_marks[(u, j)]   
                    elif (u, j) != (u, v):
                        b = np.random.randint(0, 2**63, dtype=np.int64)
                        list_marks[(u, j)] = b
                        list_marks[(j, u)] = b 
                        a = a ^ b 
                        
                list_marks[(u, v)] = a
                list_marks[(v, u)] = a  
            out[v] = min(out[v], out[u])

def compute_bridges_rand(graph):
    result.clear()
    list_marks.clear()
    for i in graph:
        if (i not in visited):
            dfs_rand(graph, i, -1, 0)

    for i in list_marks:
        if list_marks[i] == 0 and (i[0], i[1]) not in result:
            result.append(tuple(sorted(i)))
            
    visited.clear()
    ent.clear()
    out.clear()
    return result
```

```python
def compute_2bridges_rand(graph, sort_fun):
    result.clear()
    list_marks.clear()
    for i in graph:
        if (i not in visited):
            dfs_rand(graph, i, -1, 0)
    
    list_marks_sorted = defaultdict(list) 
    marks = []
    
    i=0
    for edge in list_marks:
        if i%2 == 0:
            list_marks_sorted[(tuple(sorted(edge)))] = list_marks[edge]
            if list_marks[edge] not in marks:
                marks.append(list_marks[edge])
        i=i+1
    
    if len(marks) == 0:
        return []
    
    marks = sort_fun(marks) 
    
    for i in marks:
        temp = []
        for j in list_marks_sorted:
            if list_marks_sorted[j] == i:
                temp.append(j)
        result.append(temp)
    
    visited.clear()
    ent.clear()
    out.clear()
    return result
```

```python
edges = [] 

def dfs_euclean(graph, v, parent, time):
    if v not in visited:
        visited.add(v)
    time = time + 1
    ent[v] = time
    out[v] = time
    for u in graph[v]:
        if (u==parent):
            continue
        if (u in visited):
            out[v] = min(out[v], ent[u])
        else:
            dfs_euclean(graph, u, v, time)
            edges.append(tuple(u))
            out[v] = min(out[v], out[u])
            

# Driver Code
def compute_Euler_circuit_digraph(graph):
    edges.clear()
    new_graph = defaultdict(list) 
    
    for i in graph:
        for j in graph[i]:
            for z in graph[j]:
                new_graph[(i, j)].append((j, z))
    
    for i in new_graph:
        if (i not in visited):
            dfs_euclean(new_graph, i, -1, 0)
    
    edges.reverse()
    edges.append((edges[-1][1], edges[0][0]))
    
    visited.clear()
    ent.clear()
    out.clear()
    return(edges)
```

```python
def invert_in_Zp_Euclead(p,x):
    if (gcd(p,x)!=1) or (x<=0) or (x>=p):
        return 0
    tmp=p
    a, b, c, d = 1, 0, 0, 1
    while x:
        q = p // x
        p, x = x, p % x
        a, b = b, a - b*q
        c, d = d, c - d*q
    if (b<=0): 
        c=c%tmp
    return c
```
